```python
### Text Detection Using Term Frequency–Inverse Document Frequency
## A very simple example of checking whether a piece of text is plagiarized based on data, using the SVM and TF-IDF algorithms 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# Sample data (text pairs + labels) - in Polish
data = [
    #(Plagiarism / Paraphrasing - Label 1)
    ("Kupiłem nowy samochód", "Nabyłem nowe auto", 1),
    ("On czyta książkę", "On przegląda powieść", 1),
    ("Nauczyciel wyjaśnia trudne zagadnienie", "Pedagog tłumaczy skomplikowany problem", 1),
    ("Projekt został ukończony przed czasem", "Zadanie sfinalizowano przed wyznaczonym terminem", 1),
    ("Dziewczynka biegnie przez zieloną łąkę", "Małe dziecko pędzi po trawiastym polu", 1),
    ("Firma odnotowała ogromny wzrost zysków", "Przedsiębiorstwo zarejestrowało wielki skok dochodów", 1),
    ("Wieczorem lubię oglądać dobre filmy", "W nocy chętnie śledzę ciekawe seanse filmowe", 1),
    ("Programista pisze kod w języku Python", "Developer tworzy oprogramowanie przy użyciu Pythona", 1),
    ("Jutro rano czeka nas ważny egzamin", "Kolejnego dnia rano zdajemy istotny test", 1),
    ("Zupa pomidorowa smakuje wyśmienicie", "Krem z pomidorów jest niezwykle smaczny", 1),

    #(No plagiarism / Various topics - Rating 0)
    ("Lubię jeść pizzę", "Pada dzisiaj deszcz", 0),
    ("Idę do sklepu", "Gram w piłkę", 0), 
    ("Nauczyciel wyjaśnia trudne zagadnienie", "Kawa w filiżance szybko wystygła", 0),
    ("Projekt został ukończony przed czasem", "Mój pies uwielbia aportować czerwoną piłkę", 0),
    ("Dziewczynka biegnie przez zieloną łąkę", "Ceny paliw na stacjach znowu wzrosły", 0),
    ("Firma odnotowała ogromny wzrost zysków", "Księżyc mocno świecił tej nocy", 0),
    ("Wieczorem lubię oglądać dobre filmy", "Muszę pilnie naprawić cieknący kran", 0),
    ("Programista pisze kod w języku Python", "Liście spadają z drzew na jesień", 0),
    ("Jutro rano czeka nas ważny egzamin", "Rowerzysta jechał ścieżką wzdłuż rzeki", 0),
    ("Zupa pomidorowa smakuje wyśmienicie", "Nowy smartfon ma potrójny aparat", 0)
]
```

```python
# Text separation
texts1 = [x[0] for x in data]
texts2 = [x[1] for x in data]
labels = [x[2] for x in data]

# Combining text strings into a single string
combined_texts = [t1 + " [SEP] " + t2 for t1, t2 in zip(texts1, texts2)]
```

```python
# Data train/test
X_train, X_test, y_train, y_test = train_test_split(
    combined_texts, labels, test_size=0.25, random_state=42
)

# Pipeline: TF-IDF + SVM
model = Pipeline([
    ("tfidf", TfidfVectorizer(
        ngram_range=(1, 2), 
        max_df=0.9,
        min_df=1
    )),
    ("svm", SVC(kernel="linear", probability=True))
])

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# RESULTS
print(classification_report(y_test, y_pred))
```

```python
# manual code testing
test_pair = ["Małe dziecko [SEP] Dziewczynka"]
print(model.predict(test_pair))
```
