```python
import matplotlib.pyplot as plt
import numpy as np
from math import comb

# --- Example 1: Customers buying a product ---
n1 = 100 # number of customers
p1 = 0.2 # probability of buying a product
k_values1 = np.arange(0, n1 + 1)
probabilities1 = [comb(n1, k) * (p1**k) * ((1-p1)**(n1-k)) for k in k_values1]

plt.figure(figsize=(10, 5))
plt.bar(k_values1, probabilities1, color='lightgreen', edgecolor='black')
plt.title(f'Binomial Distribution: n={n1}, p={p1} (Number of Customers)')
plt.xlabel('Number of Customers (k)')
plt.ylabel('Probability')
plt.xlim(0, 50)
mean1 = n1 * p1
std_dev1 = (n1 * p1 * (1-p1))**0.5
plt.axvline(mean1, color='red', linestyle='--', label=f'Mean = {mean1:.1f}')
plt.legend()
plt.show()

print(f"Mean of customers buying a product with sample 100 and probability 0.2 is equal to: {mean1:.2f}")
print(f"Standard deviation of buying a product with sample 100 and probability 0.2 is equal to: {std_dev1:.2f}")


# --- Example 2: A/B Test ---
n2 = 50  # number of customers in the test
pA = 0.65  # probability of clicking button A
pB = 0.35  # probability of clicking button B

# Distribution for button A using SciPy
from scipy.stats import binom

k_values2 = np.arange(0, n2 + 1)
probabilitiesA = binom.pmf(k_values2, n2, pA)

plt.figure(figsize=(10, 5))
plt.bar(k_values2, probabilitiesA, color='skyblue', edgecolor='black')
plt.title(f'Binomial Distribution: n={n2}, p={pA} (Clicks on A)')
plt.xlabel('Number of Clicks on A (k)')
plt.ylabel('Probability')
plt.xlim(20, 45)
meanA = n2 * pA
std_devA = (n2 * pA * (1-pA))**0.5
plt.axvline(meanA, color='red', linestyle='--', label=f'Mean = {meanA:.1f}')
plt.legend()
plt.show()

print(f"Mean of A/B Test with sample 50 and probability 0.65 is equal to: {meanA:.2f}")
print(f"Standard deviation of A/B Test with sample 50 and probability 0.65 is equal to: {std_devA:.2f}")
```
