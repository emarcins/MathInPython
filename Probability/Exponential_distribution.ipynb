```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameters
lambda_per_hour = 12              # 12 customers per hour
lambda_per_minute = lambda_per_hour / 60  # convert to per minute

# In scipy, the scale of the exponential distribution is 1/Î»
scale = 1 / lambda_per_minute

# Time range (0 to 10 minutes)
x = np.linspace(0, 10, 500)

# Part I - PDF
# Probability Density Function (PDF)
pdf = expon.pdf(x, scale=scale)

# Calculate the probability that a customer arrives within 10 minutes
prob_within_10_min = expon.cdf(10, scale=scale)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Probability Density Function (PDF)', color='blue')
plt.fill_between(x, pdf, where=(x <= 10), color='skyblue', alpha=0.5, label='P(X â‰¤ 10 min)')
plt.axvline(10, color='red', linestyle='--', label='10 minutes')
plt.title('Exponential Distribution of Waiting Time for a Customer\n(average 12 customers per hour)')
plt.xlabel('Waiting Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()

print(f'Probability that a customer arrives within 10 minutes under the above assumptions: â‰ˆ {prob_within_10_min:.3f}')

# Part II - CDF
# Cumulative Distribution Function (CDF)
cdf = expon.cdf(x, scale=scale)

# CDF Plot
plt.figure(figsize=(10, 6))
plt.plot(x, cdf, label='Cumulative Distribution Function (CDF)', color='green')
plt.axvline(10, color='red', linestyle='--', label='10 minutes')
plt.axhline(prob_within_10_min, color='orange', linestyle='--', label=f'P(X â‰¤ 10 min) â‰ˆ {prob_within_10_min:.3f}')
plt.title('CDF of the Exponential Distribution of Waiting Time for a Customer')
plt.xlabel('Waiting Time (minutes)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
