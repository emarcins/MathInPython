```python
# Gamma distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Distribution parameters
alpha = 3     # number of events
beta = 0.1    # inverse of the average event rate
scale = 1 / beta  # scipy uses the "scale" parameter, which is 1 / beta

# Time value range
x = np.linspace(0, 100, 500)

# Gamma distribution density
pdf = gamma.pdf(x, a=alpha, scale=scale)

# Calculate mean and mode
mean = alpha / beta
mode = (alpha - 1) / beta if alpha > 1 else 0  # mode only valid for alpha > 1, the highest probability

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f"Gamma(α={alpha}, β={beta})", color='blue')
plt.axvline(mean, color='green', linestyle='--', label=f"Mean = {mean:.1f} min")
plt.axvline(mode, color='red', linestyle='--', label=f"Mode = {mode:.1f} min")
plt.title("Gamma Distribution – Time until 3rd event")
plt.xlabel("Time (minutes)")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()
plt.show()
```
