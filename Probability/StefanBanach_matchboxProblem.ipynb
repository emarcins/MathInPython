```python
## Stefan Banach’s Matchbox Problem

"""
Imagine a person who has two pockets, each containing a matchbox with the same number of matches. 
Whenever a match is needed, the person reaches into one of the pockets at random. 
At some point, it turns out that the selected matchbox is empty.

The question is:
What is the probability that exactly k matches remain in the second box?
"""
```

```python
# Stefan Banach's matchbox problem is described by the formula:
# P(k) = \frac{\binom{2n-k}{n}}{2^{2n-k+1}}
```

```python
import math
import matplotlib.pyplot as plt

def binomial(n, k):
    return math.comb(n, k)

# Banach's matchbox problem formula
def P(n, k):
    return binomial(2*n - k, n) / (2 ** (2*n - k + 1))

# calculation of a discrete distribution 
def distribution(n):
    ks = list(range(0, n+1))
    probs = [P(n, k) for k in ks]
    
    # conversion to percentages
    probs_percent = [p * 100 for p in probs]
    
    return ks, probs_percent
```

```python
# drawing plot
def plot_distribution(n):
    ks, probs = distribution(n)
    
    plt.figure()
    plt.bar(ks, probs)
    plt.xlabel("k (number of remaining matches)")
    plt.ylabel("Probability (%)")
    plt.title(f"Distribution P(k) for n = {n}")
    plt.show()

# simulations:

# 1) 2 boxes of matches, 20 matches each
plot_distribution(20)

# 2) 2 boxes of matches, 50 matches each
plot_distribution(50)
```
