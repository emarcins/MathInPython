```python
## The Law of Large Numbers, illustrated by rolling a six-sided die

#The expected average of the sum of all rolls, divided by the number of rolls, 
#should approach 3.5 as the number of trials increases
```

```python
import random
```

```python
def roll(num_trials):
    rolls = []
    
    for t in range(num_trials):
        rolls.append(random.randint(1,6))
    return sum(rolls)/num_trials
```

```python
if __name__ == '__main__':
    expected_value = 3.5

    print(f"Expected value: {expected_value}")
    for trial in [100, 1000, 10000, 100000, 500000]:
        avg = roll(trial)
        print(f"Number of throws: {trial}. Average trials score: {avg}.")
```

```python
## Estimating the value of pi using the Law of Large Numbers
```

```python
# The program estimates the value of pi, which can be obtained by estimating the area of a circle through 
# a simulation involving the placement of random points. 
# Thanks to the Law of Large Numbers, as the number of points increases, we obtain an approximation of pi.
```

```python
import math
import random
```

```python
def estimate(total_points):
    radius = 1 
    center = (radius, radius)

    in_circle = 0
    for i in range(total_points):
        x = random.uniform(0, 2*radius)
        y = random.uniform(0, 2*radius)
        p = (x,y)
        # distance from the center of the circle
        d = math.sqrt((p[0]-center[0])**2 + (p[1]-center[1])**2)
        if d <= radius:
            in_circle += 1
    return (in_circle/total_points)*4

if __name__ == "__main__":
    for points in [10**3, 10**4, 10**5, 10**6]:
        print(f"The value of pi: {math.pi}, Estimated value ({points} points): {estimate(points)}")
```
