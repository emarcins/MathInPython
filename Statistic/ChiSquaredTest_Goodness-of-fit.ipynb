```python
# Goodness-of-fit chi-squared test

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Observed data: number of times each face appeared in 60 dice rolls
observed = np.array([8, 9, 14, 11, 10, 8])
expected = np.array([10, 10, 10, 10, 10, 10])  # uniform distribution

# 🔹 Performing the chi-square test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
df = len(observed) - 1
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)

# Test results
print("Chi-square statistic value:", chi2_stat)
print("Degrees of freedom:", df)
print("Critical value (for α = 0.05):", critical_value)
print("p-value:", p_value)

# Hypothesis testing - The H0 is that the dice is fair
if chi2_stat > critical_value:
    print("Reject the null hypothesis – the distribution may not be uniform.")
else:
    print("Do not reject the null hypothesis – the differences may be due to chance.")

# 🔹 Chi-square distribution plot
x = np.linspace(0, 25, 500)
y = stats.chi2.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Chi-square distribution (df={df})', color='blue')
plt.axvline(chi2_stat, color='red', linestyle='--', label=f'Test statistic = {chi2_stat:.2f}')
plt.axvline(critical_value, color='green', linestyle='--', label=f'Critical value (α=0.05) = {critical_value:.2f}')
plt.fill_between(x, y, where=(x > critical_value), color='green', alpha=0.3, label='Rejection region (H₀)')
plt.title("Chi-square test – distribution fit")
plt.xlabel("Chi-square statistic value")
plt.ylabel("Probability density")
plt.legend()
plt.grid(True)
plt.show()
```
