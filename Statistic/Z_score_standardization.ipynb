```python
# Z-score standardization

import numpy as np
from sklearn.preprocessing import StandardScaler

# Data â€“ height in cm
height = np.array([160, 165, 170, 175, 180])

# --- Manual Z-score calculation ---
mean_value = np.mean(height)
std_dev = np.std(height)

# Compute Z-scores manually
z_scores_manual = (height - mean_value) / std_dev

print("Mean:", mean_value)
print("Standard deviation:", std_dev)
print("------------------------------------------------------------------------------")
print("Manual Z-score:", z_scores_manual)
print("------------------------------------------------------------------------------")

# --- Z-score using Scikit-Learn ---
# Reshape the array to 2D (required by Scikit-Learn)
height_reshaped = height.reshape(-1, 1)

# Create the scaler object
scaler = StandardScaler()

# Fit and transform the data
z_scores_sklearn = scaler.fit_transform(height_reshaped)

print("Scikit-Learn Z-score:", z_scores_sklearn.flatten())
```
