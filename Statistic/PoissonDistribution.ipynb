```python
# Poisson distribution for an average of 2 typos per statistically assumed 100 pages of a book.

"""
Poisson Distribution formula:
    P(k;λ)= (e**(-λ) * λ**k) / factorial(k)
"""

from scipy.stats import poisson
import matplotlib.pyplot as plt

# Average number of typos (lambda - λ)
lambda_typos = 2

# Range of the number of typos to be analyzed (0 to 10)
k_values = range(0, 11)

# Calculating the probabilities
probs = [poisson.pmf(k, lambda_typos) for k in k_values]

# Show results
for k, p in zip(k_values, probs):
    print(f"Probability of {k} typos: {p:.4f}")

# Bar chart
plt.bar(k_values, probs, color='skyblue')
plt.xlabel("Number of typos in 100 pages of a book")
plt.ylabel("Probability")
plt.title("Poisson Distribution (λ = 2)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
