```python
# Calculating the global maximum of a single-variable function using the gradient ascent method 
# The programme also checks whether there is a solution to the equation f'(x) = 0
# and displays the analysed intermediate points on the graph
```

```python
from sympy import Derivative, Symbol, sympify, solve
from sympy.core.sympify import SympifyError
import matplotlib.pyplot as plt
```

```python
def gradient_ascent(x0, f1x, x):
    if not solve(f1x): # checks whether there is a solution to the equation f'(x) = 0
        print("It is not possible to continue; the equation {0}=0 has no solution".format(f1x))
        return None
    epsilon = 1e-6 # the value used by the algorithm to determine when to stop (it should be small)
    step_size = 1e-4
    x_old = x0 # initial value of the function
    x_new = x_old + step_size*f1x.subs({x:x_old}).evalf()

    X_traversed = []    
    while abs(x_old - x_new) > epsilon:
        X_traversed.append(x_new)
        x_old = x_new
        x_new = x_old + step_size*f1x.subs({x:x_old}).evalf()
    
    return x_new, X_traversed
```

```python
def frange(start, final, interval):

    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
    
    return numbers
```

```python
def create_plot(X_traversed, f, var):
    # Creating the plot of function
    x_val = frange(-1, 1, 0.01)
    f_val = [f.subs({var:x}) for x in x_val]
    plt.plot(x_val, f_val, 'bo')
    
    # Calculating the values of the function at all the intermediate points checked
    f_traversed = [f.subs({var:x}) for x in X_traversed]
    plt.plot(X_traversed, f_traversed, 'r.')
    plt.legend(['Function', 'Intermediate points'], loc='best')
    plt.show()
```

```python
if __name__ == "__main__":
    f = input("State the function of one variable: ")
    var = input("Specify the variable which the derivative of the function is to be calculated: ")
    var0 = float(input("Enter the initial value of the function: "))

    try:
        f = sympify(f)
    except SympifyError:
        print("The function entered is incorrect")
    else:
        var = Symbol(var)
        d = Derivative(f, var).doit()
        var_max, X_traversed = gradient_ascent(var0, d, var)
        if var_max:
            print("{0}: {1}".format(var.name, var_max))
            print("Maximum value: {0}".format(f.subs({var:var_max})))
            create_plot(X_traversed, f, var)
```
