```python
# Derivative calculator in Sympy
```

```python
from sympy import Symbol, Derivative, sympify, pprint
from sympy.core.sympify import SympifyError
```

```python
# Function template for calculating derivatives in Sympy

def derivative(f, var):
    var = Symbol(var)
    d = Derivative(f, var).doit()
    pprint(d)
```

```python
if __name__ == '__main__':
    f = input("Enter the function to be calculated: ")
    var = input("Enter the variable with respect to which the derivative is to be calculated: ")

    # error handling if the function is not saved in the correct format
    try:
        f = sympify(f)
    except SympifyError:
        print("Invalid input")
    else:
        derivative(f, var)
```
