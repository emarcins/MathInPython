```python
## The Maclaurin series is the Taylor series about the point 0
# Maclaurin series for the function f(x) = e**x

''' display the Maclaurin series for function e^x at the point 0:
        x^2   x^3   x^4       x^n
1 + x + --- + --- + --- ... + ---
        2!    3!    4!        n!

because for the function f(x) = e^x, all derivatives are equal to 1 at the point 0'''
```

```python
from sympy import Symbol, pprint, init_printing, factorial, exp, sin, series, summation, pi
```

```python
def Maclaurin_series(n):
    init_printing(order='rev-lex')

    x = Symbol('x')
    f = exp(x)
    x0 = 0

    seriesMaclaurin = 0
    
    for i in range(n):
        derivative = f.diff(x, i)
        value_at_0 = derivative.subs(x, x0)
        
        term = (value_at_0 / factorial(i)) * (x - x0)**i 
        seriesMaclaurin += term

    print("The Maclaurin series for the function f(x) = e^x:")
    pprint(seriesMaclaurin, order='rev-lex')

    series_value = seriesMaclaurin.subs({x:value_at_0})
    print(f'The value of the Maclaurin series for x = {value_at_0} is: {series_value:.15f}')

if __name__ == '__main__':
    n = int(input('Enter the number of terms in the series: '))
    Maclaurin_series(n)


print("\nTip: The higher the value, the more accurate the approximation of Euler's number (2.7182818284590...) \n(i.e., the result of the series for the function f(x) = e**x)")
```

```python
## Taylor series for the function f(x) = sinus
# Code written using an alternative method
```

```python
def taylor_sin():
    init_printing(order='rev-lex')
   
    x = Symbol('x')
    function = sin(x)
    
    num_terms = int(input('Enter the number of terms in the series: '))
    
    # compute Maclaurin series (around 0), removed 0 values
    taylor_series = function.series(x, 0, num_terms).removeO()
    
    print('\nTaylor series of sin(x):')
    pprint(taylor_series, order='rev-lex')
    
    x_value = float(input('\nEnter value of x: '))
    
    result = taylor_series.subs(x, x_value)    
    print(f'\nApproximate sin({x_value}) = {result:.8f}')


if __name__ == '__main__':
    taylor_sin()
```

```python
## Basel problem
# Code written using an alternative method, using SymPy summation

''' Basel problem solved by Leonard Euler
 1     1     1         1    π^2
--- + --- + --- ... + --- = ---
1^2   2^2   3^2       n^2    6
'''
```

```python
def approximate_basel_problem():
    n = Symbol('n')
    
    num_terms = int(input("Enter number of terms: "))
    print('\nTip: The higher the value of successive terms in the series, the more accurate the approximation')
    
    # partial sum
    partial_sum = summation(1 / n**2, (n, 1, num_terms))
    
    print(f'\nApproximation of the Basel problem with {num_terms} terms:')
    print(float(partial_sum))
    
    # compare with exact value π^2 / 6
    exact_value = (pi**2) / 6
    print('\nExact value (π^2 / 6):')
    print(exact_value.evalf())


if __name__ == '__main__':
    approximate_basel_problem()
```
