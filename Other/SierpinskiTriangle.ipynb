```python
## The script draws a Sierpinski triangle
```

```python
import random 
import matplotlib.pyplot as plt
```

```python
# transformations
def transformation_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5*x
    y1 = 0.5*y
    return x1, y1

def transformation_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5*x + 0.5
    y1 = 0.5*y + 0.5
    return x1, y1

def transformation_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5*x + 1
    y1 = 0.5*y
    return x1, y1
```

```python
# counter of probability index
def get_index(probability):
    r = random.random()
    c_probability = 0
    sum_probability = []
    for p in probability:
        c_probability += p
        sum_probability.append(c_probability)
        
    for item, sp in enumerate(sum_probability):
        if r <= sp:
            return item
    return len(probability)-1
```

```python
def transform(p):
    transformations = [transformation_1, transformation_2, transformation_3]
    probability = [1/3, 1/3, 1/3]
    # choose a random transformation function
    tindex = get_index(probability)
    t = transformations[tindex]
    x,y = t(p)
    return x,y
```

```python
def draw_sierpinski(n):
    # start at the point with coordinates (0,0)
    x = [0]
    y = [0]

    x1, y1 = 0, 0
    for i in range(n):
        x1, y1 = transform((x1, y1))
        x.append(x1)
        y.append(y1)
    return x,y
```

```python
if __name__ == "__main__":
    n = int(input("Enter the number of points in a Sierpinski triangle: "))
    x,y = draw_sierpinski(n)

    plt.plot(x, y, "o")
    plt.title(f"Sierpinski triangle consisting of {n} points")
    plt.show()
```
