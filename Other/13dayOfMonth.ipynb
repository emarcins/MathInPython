```python
import datetime
import matplotlib.pyplot as plt
import calendar

def count_13th_by_weekday():
    """
    Counts how many times the 13th day of the month falls on each weekday
    from January 1, year 1 to January 1, 2025.
    
    Returns:
        dict: A dictionary where keys are weekday numbers (0=Monday, 6=Sunday)
              and values are the counts.
    """
    # Initialize a dictionary to store counts for each weekday
    # Monday=0, Tuesday=1, ..., Sunday=6
    weekday_counts = {i: 0 for i in range(7)}

    # Define the start and end dates for our loop
    start_date = datetime.date(1, 1, 13)
    end_date = datetime.date(2025, 1, 1)

    # Loop through all dates, incrementing by month
    current_date = start_date
    while current_date <= end_date:
        # Get the weekday number (0=Monday, 6=Sunday)
        weekday_number = current_date.weekday()
        # Increment the counter for that specific weekday
        weekday_counts[weekday_number] += 1

        # Calculate the next month's 13th day
        year = current_date.year
        month = current_date.month
        
        # Move to the next month
        if month == 12:
            year += 1
            month = 1
        else:
            month += 1
        
        # Check if the 13th of the new month is a valid date
        try:
            current_date = datetime.date(year, month, 13)
        except ValueError:
            break

    return weekday_counts

def visualize_results(weekday_counts):
    """
    Creates a bar chart to visualize the number of times the 13th of the month
    fell on each day of the week.
    
    Args:
        weekday_counts (dict): A dictionary with weekday counts.
    """
    # Create a list of weekday names
    weekday_names = [calendar.day_name[i] for i in range(7)]

    # Extract the counts from the dictionary
    counts = [weekday_counts[i] for i in range(7)]

    # Bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(weekday_names, counts, color='skyblue')
    plt.xlabel('Weekday')
    plt.ylabel('Number of occurrences on the 13th day of the month')
    plt.title('Occurrence of the 13th day of the month on specific days of the week (from 01.01.0001 to 01.01.2025)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

if __name__ == '__main__':
    counts = count_13th_by_weekday()
    
    # Map weekday numbers back to names for a more readable printout
    weekday_names = [calendar.day_name[i] for i in range(7)]
    
    print("Summary of occurrences on the 13th day of the month from 1 AD to 2025 AD:")
    for i in range(7):
        print(f"{weekday_names[i]}: {counts[i]} times")
        
    visualize_results(counts)
```
