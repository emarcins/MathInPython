```python
import matplotlib.pyplot as plt
import math
```

```python
def draw_graph(x, y):
    plt.plot(x,y)
    plt.xlabel('x-coordinate')
    plt.ylabel('y-coordinate')
    plt.title('throwing ball trajectory')

def float_range(start,final, interval):
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
    return numbers
```

```python
def draw_trajectory(u, theta):
    # Calculate the angle with the Earth(theta).
    # Convert degrees to radians.
    theta = math.radians(theta)
    # The value of Earth's acceleration.
    g = 9.8

    # Flight time.
    # To simplify calculations, the horizontal component of velocity remains unchanged, 
    # with the vertical component is denoted by the formula: u * sin * theta.

    # The formula for the total flight time of the ball corresponds to the moment when the ball reaches its highest point(t-peak) 
    # multiplied by 2 (from ascent to subsequent descent after exceeding t-peak) and divided by the acceleration due to gravity (g).
    # It means: t_flight = 2t-peak = 2*(u*sin(theta)/g)
    # u - initial speed. 
    t_flight = 2*u*math.sin(theta)/g

    # Determination of time points.
    intervals = float_range(0, t_flight, 0.001)

    # x and y coordinate
    x = []
    y = []
    for t in intervals:
        # ball thrown horizontally
        x.append(u*math.cos(theta)*t)
        # ball thrown vertically
        y.append(u*math.sin(theta)*t - 0.5*g*t*t)

    draw_graph(x, y)
```

```python
if __name__ == '__main__':
    try: 
        u = float(input('Enter the initial velocity (m/s): '))
        theta = float(input('Enter the angle of the throw (in degrees): '))
    except ValueError:
        print('Incorrect input data')
    else:
        draw_trajectory(u, theta)
        plt.show()
```

```python
## Comparison of trajectories of throws the with different initial velocities (20, 40 and 60 m/s) and 45-degree angle.

if __name__ == '__main__':
    u_list = [20, 40, 60] # list with 3 different initial speeds
    theta = 45 # ball thrown at a 45-degree angle

    for u in u_list:
        draw_trajectory(u, theta)
    
    plt.legend(['20m/s', '40m/s', '60m/s'])
    plt.title('Comparison of throwing ball trajectory for 20, 40 and 60m/s with 45-degree angle')
    plt.show()
```
