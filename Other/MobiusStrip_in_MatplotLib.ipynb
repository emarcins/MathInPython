```python
# MÃ¶bius strip

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.tri import Triangulation
```

```python
# Strip creating
theta = np.linspace(0, 2 * np.pi, 30)  # lenght
w = np.linspace(-0.25, 0.25, 8)  # width
w, theta = np.meshgrid(w, theta)

# Rotation
phi = 0.5 * theta
```

```python
### Radius in the x-y plane
r = 1 + w * np.cos(phi) # distance of each point from the centre of the coordinate system

# cordinate:
x = np.ravel(r * np.cos(theta))
y = np.ravel(r * np.sin(theta))
z = np.ravel(w * np.sin(phi))
```

```python
# Triangulation
tri = Triangulation(np.ravel(w), np. ravel(theta))
```

```python
# Plot
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap='Blues', linewidths=0.2);

ax.set_xlim(-1,1); ax.set_ylim(-1,1); ax.set_zlim(-1,1)
#ax.axis('off');
```
