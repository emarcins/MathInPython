```python
## Research on the relationship between the Fibonacci sequence and the golden ratio 
# The golden ratio is equal  φ = 1,6180339887...
```

```python
import matplotlib.pyplot as plt
```

```python
def fibo(n):
    # for n <= 2
    if n == 1:
        return [1]
    if n == 2:
        return [1,1]

    # for n > 2
    a = 1
    b = 1

    series = [a, b]
    for i in range(n):
        c = a + b
        series.append(c)
        a = b
        b = c
        
    return series
    
def plot_ratio(series):
    ratios = []
    for i in range(len(series)-1):
        ratios.append(series[i+1]/series[i])

    last_ratio = series[-1] / series[-2]
    # Result
    print(f"Fibonacci sequence: {series}")
    print(f"An approximation of the golden ratio for the {len(series)} terms:")
    print(f"{series[-1]} / {series[-2]} = {last_ratio:.10f}")

    # Golden ratio 
    golden_ratio = (1 + 5**0.5) / 2
    print(f"the difference from φ: {abs(last_ratio - golden_ratio):.10f}")

    plt.plot(ratios)
    plt.title("The Fibonacci sequence and the golden ratio coefficient")
    plt.ylabel("Coefficient")
    plt.xlabel("Number")
    plt.axhline(y=golden_ratio, color='r', linestyle='--', 
                label=f'Golden ratio φ ≈ {golden_ratio:.6f}')
    plt.legend()
    plt.show()
```

```python
if __name__ == "__main__":
    num = 15 # number of terms in the Fibonacci sequence
    series = fibo(num)
    plot_ratio(series)
```
